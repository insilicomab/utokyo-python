```python
class HelloForEver:
    def readline(self):
        return 'Hello.\n'
```

```python
class HelloFile:
    def __init__(self, n):
        self.n = n
    def readline(self):
        if self.n == 0:
            return ''
        self.n = self.n - 1
        return 'Hello.\n'
```

```python
class HelloFile(HelloForEver):
    def __init__(self, n):
        self.n = n
    def readline(self):
        if self.n == 0:
            return ''
        self.n = self.n - 1
        return super().readline()
```

```python
class HelloFileIterator(HelloFile):
    def __iter__(self):
        return self
    def __next__(self):
        line = self.readline()
        if line == '':
            raise StopIteration
        return line
```

## 練習

`'Hello.\n'` ではなくて、初期時に指定された文字列を繰り返し返すように、
新たなクラス `StringFileIterator` を定義してください。

`StringFileIterator` は `HelloFileIterator` を継承し、
初期化メソッドには、文字列と回数を指定します。

```python
class StringFileIterator(HelloFileIterator):
    def __init__(self, s, n):
        self.s = s
        self.n = n
    
    def readline(self):
        if self.n == 0:
            return ''
        self.n = self.n - 1
        return self.s
```

上のセルで解答を作成した後、以下のセルを実行し、実行結果が `True` になることを確認してください。

```python
f = StringFileIterator('abc', 3)
print(list(f) == ['abc','abc','abc'])
```
