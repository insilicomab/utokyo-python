## 練習

1. 数値 `x` の絶対値を求める関数 `absolute(x)` を定義してください。
   （Pythonには `abs` という組み込み関数が用意されていますが。）
2. `x` が正ならば 1、負ならば -1、ゼロならば 0 を返す `sign(x)` という関数を定義してください。

定義ができたら、その次のセルを実行して、`True` のみが表示されることを確認してください。

```python
def absolute(x):
  if x < 0:
    return -1 * x
  else:
    return x
```

```python
def sign(x):
  if x > 0:
    return 1
  elif x < 0:
    return -1
  else:
    return 0
```

```python
print(absolute(5) == 5)
print(absolute(-5) == 5)
print(absolute(0) == 0)
print(sign(5) == 1)
print(sign(-5) == -1)
print(sign(0) == 0)
```

## 練習の解答

```python
def absolute(x):
    if x<0:
        return -x
    else:
        return x
```

```python
def sign(x):
    if x<0:
        return -1
    if x>0:
        return 1
    return 0
```
